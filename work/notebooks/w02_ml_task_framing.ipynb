## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
# My Lane as an ML Task

Previously Selected Lane: Refresh / Content Opportunity Scoring

Type of ML Task: Ranking/Scoring

This task seeks to give priority scores for content pages according to the signals seen from search and engagement in order for content teams to be able to look at the pages with the best potential first. Unlike other tasks, the model will not predict a Yes/No outcome but a priority score/ranking.
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
# Target or Proxy

There is no direct column that says whether a page should be refreshed. Instead, the project uses observable search and engagement signals as a proxy for refresh priority. Pages showing declining performance or weaker engagement are expected to receive higher priority for human review.

The model supports prioritization rather than making a final decision automatically.
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
# Success Metric

The main success metric would be Precision@K since the model returns an ordered list of pages to be reviewed. Higher value of Precision@K means that the proportion of useful pages on the top of the list is higher, meaning that reviewers' efforts will not be wasted.
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
# Why ML Is Superior to a Static Rule

The static rule is generally based on one or two signals, for example, low click-through rate or decreasing number of impressions. Nevertheless, performance of the content is determined by a range of signals, such as search visibility, user engagement, age of the content, and its ranking.

ML algorithm allows discovering patterns among several signals simultaneously and provides a more valuable ranking of priorities than a static rule does. Recommendations should be still considered as aids for making decisions.
```
